```python
import pickle
import pandas as pd
import numpy as np
```

```python
#script to load the pipeline
def model_loader(model):
    with open(model, "rb") as f:
        pipeline = pickle.load(f)
    return pipeline   
a = model_loader("pipeline_v1.bin") 
```

```python
record = {
    "lead_source": "paid_ads",
    "number_of_courses_viewed": 2,
    "annual_income": 79276.0
}
```

```python
X_new = [record]
probabilities = a.predict_proba(X_new)[0]

convert_prob = probabilities[1]

print(f"Probability that this lead will convert: {convert_prob:.3f}")
```
